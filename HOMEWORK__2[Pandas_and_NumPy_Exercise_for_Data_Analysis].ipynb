# **NumPy Array Exercises**

```python
# Create a Numpy array filled with all zeros

import numpy as np

zeros_array = np.zeros((2,3), dtype = 'int')
print (zeros_array)
```

```python
# Empty NumPy array

empty_array = np.empty((3, 3))
print(empty_array)
```

```python
# Full NumPy array

full_array = np.full((2, 3), 6)
print(full_array)
```

```python
# Numpy array filled with all ones

ones_array = np.ones((2, 3), dtype = 'int')
print(ones_array)
```

```python
# Combining a one and a two-dimensional NumPy Array

one_dim_array = np.array([1, 2, 3])
two_dim_array = np.array([[4, 5, 6], [7, 8, 9]])

arr_add = np.vstack((one_dim_array, two_dim_array))
print (arr_add)
```

```python
#Counts the number of non-zero values in the array

array = np.array ([2, 3, 0, 5, 0, 3, 5, 4, 0, 8, 0])

c = np.count_nonzero(array)
print('Total numbers of non-zero values in array: ', c)
```

# **NumPy Matrix Exercises**

```python
import numpy as np

#Maximum value from given matrix

matrix = np.matrix('[10, 32, 63; 40, 15, 96; 47, 81, 59]')

mtx = matrix.max()

print(mtx)
```

```python
#Minimum value from given matrix

matrix = np.matrix('[10, 32, 63; 40, 5, 96; 47, 81, 59]')
mtx = matrix.min()
print(mtx)
```

```python
#Find the number of rows and columns

matrix = np.matrix('[10, 32, 63; 40, 5, 96; 47, 81, 59]')
print(matrix.shape)
```

```python
#Matrix Multiplication in NumPy

c = np.matrix([[10, 32],
              [63, 40]])
d = np.matrix([[5, 96],
              [28, 59]])

print(c * d)
```

```python
#Adding and Subtracting Matrices in Python

c = np.matrix([[10, 32], [63, 40]])
d = np.matrix([[5, 96], [28, 59]])

addition_result = np.add(c, d)
subtraction_result = np.subtract(c, d)

print ('Matrix adding: ')
print (addition_result)

print ('Matrix subtact: ')
print (subtraction_result)
```

```python
#Add row/columns in numpy array

import numpy as np

# Adding rows
c = np.array([[10, 32], [63, 40]])
d = np.array([[50, 96], [28, 59]])
combined_rows = np.vstack((c,d))
print("Combined Rows:")
print(combined_rows)

# Adding columns
e = np.array([[27], [81], [55], [31]])
combined_cols = np.hstack((combined_rows, e))
print("Combined Columns:")
print(combined_cols)
```

```python
#Count the frequency of unique values in NumPy array

import numpy as np

array1 = np.array ([6, 15, 27, 6, 30, 15])

len(np.unique(array1))
```

```python
#Inverse a matrix

matrix = np.matrix ([[10, 32], [63, 40]])
print ('Inverse matrix')

print (matrix.I)
```

```python
#Calculate the determinant of a matrix

matrix = np.matrix ([[10, 32], [63, 40]])
print (np.linalg.det(matrix))
```

```python
#Convert covariance matrix to correlation matrix using Python

cov_matrix = np.matrix([[2, 1], [1, 2]])
correlation_matrix = np.corrcoef(cov_matrix)
print("Correlation Matrix:")
print(correlation_matrix)
```

# **NumPy Indexing Exercises**

```python
import numpy as np

#Replace NumPy array elements with condition

array = np.array([1, 2, 3, 4, 5])
condition = array < 3

array[condition] = 0
print(array)
```

```python
#Return the indices(indexs')  of elements where the given condition is satisfied

array = np.array ([6, 15, 27, 6, 30, 15])

indices = np.where(array > 12)
print(indices)
```

```python
#Replace negative value with zero in numpy array

array = np.array([-1, 2, -3, 4, -5])
condition = array < 0
array[condition]=0

print (array)
```

# **Pandas Daraftame Columns Exercises**

```python
#Create a pandas column using for loop


import pandas as pd

df =  pd.DataFrame({'Name': ['X', 'Y', 'Z'],'Last name': ['A','B','C'],
                    'Gender': ['Male', 'Female', 'Male'], 'Marks': [12, 52, 36]})


result = []
for value in df["Marks"]:
    if value >= 33:
        result.append("Pass")
    elif value < 0 and value > 100:
        result.append("Invalid")
    else:
        result.append("Fail")

df["Result"] = result
df
```

```python
#How to get column names in Pandas dataframe

import pandas as pd

df =  pd.DataFrame({'Name': ['X', 'Y', 'Z'],'Last name': ['A','B','C'],
                    'Gender': ['Male', 'Female', 'Male'], 'Marks': [12, 52, 36]})

for i in df.columns:
    print(i)
```
